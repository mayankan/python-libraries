```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_scatter_plot(x_input, y_input):
    """
    Description:
        Plots Pie Chart using given x and y inputs with different sizes.
    Parameter:
        x_input: List with values to be plotted as X-Axis.
        y_input: List with values to be plotted on Y-Axis.
    Return:
        None.
    """
    areas= np.pi * np.random.randint(5,15,25) ** 2
    plt.scatter(x_input, y_input, s = areas, alpha = 0.85)
```

```python
def main():
    sample_x = np.random.rand(25)
    sample_y = np.random.rand(25)
    plot_scatter_plot(sample_x, sample_y)
    plt.show()
```

```python
if __name__ == "__main__":
    main()
```
