```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
def main():
    data = sns.load_dataset("iris")
    sns.violinplot(x = "species", y = "sepal_length", data = data)
```

```python
if __name__ == "__main__":
    main()
```
