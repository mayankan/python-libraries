```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
def main():
    data = sns.load_dataset("titanic")
    sns.pointplot(x = "sex", y = "survived", data = data, estimator = np.cov)
```

```python
if __name__ == "__main__":
    main()
```
