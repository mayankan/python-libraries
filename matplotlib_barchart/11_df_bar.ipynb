```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def plot_bar_chart(data_frame):
    """
    Description:
        Plots Bar Chart with using given pandas data frame inputs.
    Parameter:
        data_frame: Pandas Data Frame with values to be plotted on X-Axis and Y-Axis.
    Return:
        None/
    """
    data_frame.plot.bar()
    plt.grid(which='major',linestyle='-',color='blue')
    plt.grid(which='minor',linestyle=":",color='black')
```

```python
def set_labels_title_show():
    """
    Description:
        Calls Show command of Pyplot after Setting the X and Y Axis Labels, Title and Grid on of the plot.
    Parameter:
        None.
    Return:
        None.
    """
    # Turn on the grid
    plt.minorticks_on()
    plt.title("Popularity of programming languages")
    plt.xlabel("Programming languages")
    plt.ylabel("Popularity")
    plt.show()
```

```python
def main():
    sample_df = pd.read_csv('df.csv')
    plot_bar_chart(sample_df)
    set_labels_title_show()
```

```python
if __name__ == "__main__":
    main()
```
