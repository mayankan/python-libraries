```python
import matplotlib.pyplot as plt
```

```python
def plot_bar_chart(x_input, y_input, y_width):
    """
    Description:
        Plots Bar Chart with no bottom margin of each bar using given x and y inputs.
    Parameter:
        x_input: List with values to be plotted on X-Axis.
        y_input: List with values to be plotted on Y-Axis.
    Return:
        None
    """
    y_pos = [0, 1, 2, 3, 4, 5]
    plt.bar(y_pos, y_input, width = y_width)
    plt.grid(which='major',linestyle='-',color='blue')
    plt.grid(which='minor',linestyle=":",color='black')
    plt.minorticks_on()
    plt.xticks(y_pos, x_input)
    plt.minorticks_on()
```

```python
def set_labels_title_show():
    """
    Description:
        Calls Show command of Pyplot after Setting the X and Y Axis Labels, Title and Grid on of the plot.
    Parameter:
        None.
    Return:
        None.
    """
    # Turn on the grid
    plt.minorticks_on()
    plt.title("Popularity of programming languages")
    plt.xlabel("Programming languages")
    plt.ylabel("Popularity")
    plt.show()
```

```python
def main():
    languages = ['Java','Python','PHP','JavaScript','C#','C++']
    popularity = [22.2,17.6,8.8,8,7.7,6.7]
    width=[0.5,0.8,0.6,0.7,0.4,0.25]
    plot_bar_chart(languages, popularity, width)
    set_labels_title_show()
```

```python
if __name__ == "__main__":
    main()
```
