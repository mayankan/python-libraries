```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def draw_subplots(x_input, y_input):
    """
    Description:
        Shows three matplotlib sub line plots using given x and y axis points.
    Parameter:
        x_input: Numpy array with values to be plotted on X-Axis.
        y_input: Numpy array with values to be plotted on Y-Axis.
    Return:
        None.
    """
    fig = plt.figure(figsize=(18, 6))

    ax = fig.add_subplot(131)

    x_limit = "{:.2f}, {:.2f}".format(ax.get_xlim()[0], ax.get_xlim()[1])
    y_limit = "{:.2f}, {:.2f}".format(ax.get_ylim()[0], ax.get_ylim()[1])
    ax.set_title(f"Full view: {x_limit}, {y_limit}")
    ax.plot(x_input, color='blue', label='Sine wave')
    ax.plot(y_input, color='black', label='Cosine wave')

    ax2 = fig.add_subplot(132)

    ax2.set_xlim([25, 50])
    x2_limit = "{:.2f}, {:.2f}".format(ax2.get_xlim()[0], ax2.get_xlim()[1])
    y2_limit = "{:.2f}, {:.2f}".format(ax2.get_ylim()[0], ax2.get_ylim()[1])
    ax2.set_title(f"Truncated X view: {x2_limit}, {y2_limit}")
    ax2.plot(x_input, color='blue', label='Sine wave')
    ax2.plot(y_input, color='black', label='Cosine wave')

    ax3 = fig.add_subplot(133)

    ax3.set_ylim([-1, 0])
    x3_limit = "{:.2f}, {:.2f}".format(ax3.get_xlim()[0], ax3.get_xlim()[1])
    y3_limit = "{:.2f}, {:.2f}".format(ax3.get_ylim()[0], ax3.get_ylim()[1])
    ax3.set_title(f"Truncated Y view: {x3_limit}, {y3_limit}")
    ax3.plot(x_input, color='blue', label='Sine wave')
    ax3.plot(y_input, color='black', label='Cosine wave')

    plt.show()
```

```python
def main():
    # Creating a numpy array from 0 to 10 with increaments of 0.1
    array = np.arange(0, 10, 0.1)
    x = np.sin(array)
    y = np.cos(array)
    draw_subplots(x, y)
```

```python
if __name__ == "__main__":
    main()
```
