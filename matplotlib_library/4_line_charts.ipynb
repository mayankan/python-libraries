```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
def draw_multiple_line(data):
    """
    Description:
        Shows a matplotlib line plot using given x and y axis points.
    Parameter:
        data: Pandas Data Frame with values to be plotted on both axis.
    Return:
        None.
    """
    plt.plot(data)
    plt.xlabel("Date")
    plt.ylabel("Income")
    plt.legend(data.columns)
```

```python
def add_title_show():
    """
    Description:
        Calls Show command of Pyplot after Setting the Title of the plot.
    Parameter:
        None.
    Return:
        None.
    """
    plt.title("Financial Data of Alphabet Inc.")
    plt.show()
```

```python
def main():
    df=pd.read_csv("fdata.csv")
    df.set_index('Date',inplace=True)
    draw_multiple_line(df)
    add_title_show()
```

```python
if __name__ == "__main__":
    main()
```
