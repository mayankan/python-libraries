```python
import matplotlib.pyplot as plt
```

```python
def draw_line_legend(x_inputs, y_inputs, line_label):
    """
    Description:
        Shows a matplotlib line plot using given x and y axis points with given line label on legend.
    Parameter:
        x_inputs: List with values to be plotted on X-Axis.
        y_inputs: List with values to be plotted on Y-Axis.
        line_label: Legend Label for Line.
    Return:
        None.
    """
    plt.plot(x_inputs, y_inputs, label = line_label)
```

```python
def add_title_legend_label_show():
    """
    Description:
        Calls Show command of Pyplot after Setting the X and Y Axis Labels, Title and Legend of the plot.
    Parameter:
        None.
    Return:
        None.
    """
    plt.xlabel("X-Axis")
    plt.ylabel("Y-Axis")
    plt.title("Display Legends")
    plt.legend()
    plt.show()
```

```python
def main():
    x = [1,2,3,4,5]
    y1 = [9,8,7,6,5]
    y2 = [4,3,2,1,1]
    y3 = [0,0,9,0,0]
    draw_line_legend(x, y1, "First Line")
    draw_line_legend(x, y2, "Second Line")
    draw_line_legend(x, y3, "Third Line")
    # Calls function to set X and Y labels, title and add legends to Plot.
    add_title_legend_label_show()
```

```python
if __name__ == "__main__":
    main()
```
