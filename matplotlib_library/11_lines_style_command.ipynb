```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def line_styles_plot(x_input):
    """
    Description:
        Shows a matplotlib plot with three lines having different styles using given x and y axis points, squaring it and cubing it.
    Parameter:
        x_input: List with values to be plotted on X-Axis.
    Return:
        None.
    """
    plt.plot(x_input, x_input, "--", x_input, x_input ** 2, "-.", x_input, x_input ** 3, ':')
    plt.show()
```

```python
def main():
    x = np.array([1,2,3,4,5])
    line_styles_plot(x)
```

```python
if __name__ == "__main__":
    main()
```
