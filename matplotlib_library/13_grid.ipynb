```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def draw_grid(data_frame):
    """
    Description:
        Shows matplotlib Line plot with Grid Lines using given Pandas Dataframe.
    Parameter:
        data_frame: Pandas Data Frame with values to be plotted on X-Axis and Y-Axis.
    Return:
        None.
    """
    plt.plot(data_frame, 'g')
    # Add Grid Lines with cyan color, linestyle dashes and line width 0.5mm.
    plt.grid(color = 'c', ls = '-', lw = .5)
```

```python
def add_title_legend_label_show():
    """
    Description:
        Calls Show command of Pyplot after Setting the X and Y Axis Labels, Title and Legend of the plot.
    Parameter:
        None.
    Return:
        None.
    """
    plt.xlabel("Date")
    plt.ylabel("Close")
    plt.title("Closing value of Alphabet lnc.")
    plt.legend()
    plt.show()
```

```python
def main():
    df = pd.read_csv("fdata.csv")
    df.set_index('Date',inplace=True)
    draw_grid(df)
    add_title_legend_label_show()
```

```python
if __name__ == "__main__":
    main()
```
