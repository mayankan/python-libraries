```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def draw_grid_ticks_off(data_frame):
    """
    Description:
        Shows matplotlib Line plot with Grid Lines ticks turned off using given Pandas Dataframe.
    Parameter:
        data_frame: Pandas Data Frame with values to be plotted on X-Axis and Y-Axis.
    Return:
        None.
    """
    fig = plt.gca()
    fig.plot(data_frame,'g-o')
    fig.set_xticklabels([])
    fig.set_yticklabels([])
    fig.grid(which='major', color='black',linestyle='-')
    fig.grid(which='minor', color='gray', linestyle='--')
```

```python
def add_title_legend_label_show():
    """
    Description:
        Calls Show command of Pyplot after Setting the X and Y Axis Labels, Title and Legend of the plot.
    Parameter:
        None.
    Return:
        None.
    """
    plt.xlabel("Date")
    plt.ylabel("Close")
    plt.title("Closing value of Alphabet lnc.")
    plt.legend()
    plt.show()
```

```python
def main():
    df = pd.read_csv("fdata.csv")
    df.set_index('Date',inplace=True)
    draw_grid_ticks_off(df)
    add_title_legend_label_show()
```

```python
if __name__ == "__main__":
    main()
```
