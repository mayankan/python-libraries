```python
# Importing required libraries
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def draw_user_line(x_input, y_input):
    """
    Description:
        Shows a matplotlib line plot using given x and y axis points.
    Parameter:
        x_input: List with values to be plotted on X-Axis.
        y_input: List with values to be plotted on Y-Axis.
    Return:
        None.
    """
    plt.plot(x_input, y_input)
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.title("Line (User axis)")
    plt.show()
```

```python
def main():
    df = pd.read_table("line_inputs.txt", delim_whitespace=True, names=('x_points', 'y_points'))
    draw_user_line(df['x_points'], df['y_points'])
```

```python
if __name__ == "__main__":
    main()
```
